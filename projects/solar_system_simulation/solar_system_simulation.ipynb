# Solar System Simulation

## Object-Oriented Programming Project

This project models and animates a simple solar system using object-oriented programming.

### Concepts Used
- classes and objects
- constructors (`__init__`)
- `self`
- accessor methods
- mutator methods
- `__str__`
- animation with Turtle graphics

### Files Used
- `planet_class.py`
- `sun_class.py`
- `solar_system_class.py`

```python
from planet_class import Planet
from sun_class import Sun
from solar_system_class import SolarSystem
```

```python
# Create the sun object
sun = Sun(0, 0, 14, "yellow")

# Create the solar system object
solar_system = SolarSystem(sun)
```

```python
# Create planet objects
mercury = Planet("Mercury", 70, 6, "gray", 0, 2.8)
venus = Planet("Venus", 120, 7, "orange", 90, 2.0)
earth = Planet("Earth", 180, 8, "blue", 180, 1.4)
mars = Planet("Mars", 240, 7, "red", 270, 1.0)
```

```python
# Add the planets to the solar system
solar_system.add_planet(mercury)
solar_system.add_planet(venus)
solar_system.add_planet(earth)
solar_system.add_planet(mars)
```

```python
# Demonstrate __str__
print(sun)
print(earth)

# Demonstrate accessor
print("Earth speed before change:", earth.get_speed())

# Demonstrate mutator
earth.set_speed(2.2)

# Confirm the change
print("Earth speed after change:", earth.get_speed())
```

```python
solar_system.show_data()
```

```python
solar_system.draw_orbits()
```

```python
solar_system.animate()
```

## Reflection

This project separates the main program from the class definitions.

- `Planet` represents individual planets
- `Sun` represents the sun
- `SolarSystem` manages the whole system

This structure is closer to how real software projects are organized.
